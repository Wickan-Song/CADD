# **常用的计算类型**

## **势能面**

- 分子体系总能量由动能和势能组成
- 分子势能：指的是在无外场时的单个分子体系的势能，即分子内部运动的势能
- 势能曲线：一维势能函数的图像被称为势能曲线
- 势能面（Potential Energy Surface）:一个体系中能量的变化反应为多维表面表示，将参数多于一个的势能图像称为势能面，势能面上每一个点对应一个特定的分子几何构型，势能面上有一些特殊性质的重要点；全局最大值，全局最小值，局域极小值以及鞍点
- 鞍点：鞍点为势能面上能量对坐标的一阶导为0的点，他是在一个方向上具有极大值，而在其他方向上具有极小值的点，一般来说，鞍点为连接两个极小值的过渡态

## **单点能**

- **单点能计算（single point energy）**：是指对于给定分子的几何构象或构象进行能量及其相关性质的计算，在势能面上只是一个点，因此叫单点能计算

## **几何优化**

- **定义**：分子在三维空间中趋向无张力低能量稳定构象的过程。分子几何构象变化对能量影响大，确定分子低能量构形（即分子势能面上的极小值）的过程就是几何优化，本质是数学上求解极值
- **原理**：从初始构型开始，计算能量梯和梯度，沿能量下降最快的方向确定下一步方向和步长

## **热力学和结构性质** 

- 分子的能量为体系的**势能和动能总和**，**势能**由体系的电子能量和核排斥能组成。
- **生成热** ：在一定温度下和压力下，由最稳定的单质生成1mol纯物质的热效应（从总能量中减去原子化热和原子的电离能）
- **立体能**：分子力学使用力场计算热力学性质，分子力学程序对于体系的任意构象课计算出能量值，该能量为体系的**立体能** 对应在体系在无任何张力状态下所对应的零点能的相对能量
- **径向分布函数**：（radial distribution functions RDF）通常指的是给定某个粒子的坐标，计算其他例子在空间的分布概率，即离给定例子的远近程度。一个以原子核为中心，半径为 r ，微单位厚度为dr （dr：在半径方向上的微小厚度或微小距离增量）的薄球壳内的体积为4πr^2dr 核外电子出现在该球壳内的概率用 g（r,r'）

## **电离势**（ionization potential）

- **含义**：指基态的气态原子或气态离子失去一个电子所需要的最小能量，衡量原子或分子失去电子的难易程度
- **意义**：研究生物分子与其他分子相互作用时，涉及电子转移的过程，有助于判断氧化-还原反应

## **偶极矩**（dipole moment）

- **含义**：衡量分子极性大小的物理量，由分子中电荷分布不均匀导致正，负电荷中心不重合产生
- **公式**：μ=q×d （其中μ表示偶极矩，q为正、负电荷中心所带的电荷量，d是正、负电荷中心之间的距离）

## **静电势**（electrostatic potential）

- **定义**：将正电荷从分子周围口昂见某点处r移到无穷远处电场力所做的功
- **静电相互作用能**：位于某一点r处的电荷与静电势的成绩

## **布局分析**
- **定义**：布局分析旨在将分子中的电子分布进行划分和定量分析，确定电子在不同原子，化学键或轨道上的分布情况从而深入了解分子的电子结构，化学键性质以及分的化学活性等
- **方法及原理**：
  1. **Mulliken布局分析**：
     **原理**：基于分子轨道理论，将分子轨道波函数（描述分子中电子运动状态的函数）表示为原子轨道的线性组合，通过计算原子轨道在分子轨道中的贡献来分配电子
  2. **自然布局分析（NPA）**：
     **原理**：在Mulliken布局分析的基础上进行改进，考虑了原子轨道的自然杂化情况，能更准确地反映原子真实电子分布
